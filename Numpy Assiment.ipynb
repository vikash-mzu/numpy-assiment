```python
Q1. Is there any difference in the data type of variables list_ and array_list? If there is, then write a code to print the data types of both the variables.
Yes, there is a difference between the data types of list_ and array_list. list_ is a Python list, while array_list is a NumPy array. You can print their data types using the following code:


import numpy as np

list_ = ['1', '2', '3', '4', '5']
array_list = np.array(object=list_)

# Print the data types
print(f"Data type of list_: {type(list_)}")
print(f"Data type of array_list: {type(array_list)}")
Q2. Write a code to print the data type of each and every element of both the variables list_ and array_list.
You can loop through both list_ and array_list to print the data types of each element.

# Data type of elements in list_
print("Data type of each element in list_:")
for element in list_:
    print(type(element))

# Data type of elements in array_list
print("Data type of each element in array_list:")
for element in array_list:
    print(type(element))
Q3. Considering the following changes in the variable array_list:
python
Copy code
array_list = np.array(object=list_, dtype=int)
Will there be any difference in the data type of the elements present in both the variables list_ and array_list? If so, then print the data types of each and every element present in both the variables list_ and array_list.

Yes, there will be a difference. After specifying dtype=int, the elements in array_list will be converted to integers, while the elements in list_ will still be strings.


# Change the dtype of array_list to int
array_list = np.array(object=list_, dtype=int)

# Print the data type of elements in list_
print("Data type of each element in list_:")
for element in list_:
    print(type(element))

# Print the data type of elements in array_list
print("Data type of each element in array_list:")
for element in array_list:
    print(type(element))
Q4. Write a code to find the following characteristics of the variable num_array:
(i) shape (ii) size


import numpy as np

num_list = [[1, 2, 3], [4, 5, 6]]
num_array = np.array(object=num_list)

# (i) Shape of num_array
print(f"Shape of num_array: {num_array.shape}")

# (ii) Size of num_array
print(f"Size of num_array: {num_array.size}")
Q5. Write a code to create a NumPy array of a 3x3 matrix containing only zeros, using a NumPy array creation function.

# Create a 3x3 matrix containing zeros
zero_matrix = np.zeros((3, 3))
print(zero_matrix)
Q6. Create an identity matrix of shape (5,5) using NumPy functions.

# Create a 5x5 identity matrix
identity_matrix = np.identity(5)
print(identity_matrix)
```
